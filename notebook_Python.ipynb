```python
# Identificar o separador ("sep") dos arquivos CSV

# O cdigo Shell Script abaixo tem como objetivo identificar o separador utilizado 
# em arquivos CSV de um diretrio especfico. Ele percorre cada arquivo CSV no 
# diretrio e realiza testes para encontrar o separador, usando uma lista de 
# possveis delimitadores.  Ao encontrar um separador no arquivo, ele exibe essa 
# informao. Essa abordagem  til para entender a estrutura dos arquivos CSV e 
# facilitar o processamento dos dados posteriormente.

%%shell

caminho="./"

for arquivo_csv in $caminho*.csv; do
    echo "Arquivo: $arquivo_csv"

    delimitadores=("," ";" "\t" "|")

    for delimitador in "${delimitadores[@]}"; do
        resultado=$(head -n 1 "$arquivo_csv" | grep -o "$delimitador" | wc -l)
        
        if [ "$resultado" -gt 0 ]; then
            echo "Separador: $delimitador"
            break
        fi
    done

    echo "-------------------------------------"
done
```

**2.6 Leitura de mltiplos arquivos CSV em dataframes individuais, atravs de um loop (for)**

Uma das tarefas mais comuns na anlise de dados  a leitura de arquivos CSV e sua transformao em dataframes do Pandas. Nesse sentido, a leitura de mltiplos arquivos CSV em dataframes individuais atravs de um loop (for)  uma tcnica muito til e eficiente em Python. Com essa tcnica,  possvel ler e manipular grandes quantidades de dados de forma organizada e automatizada, o que  fundamental para a anlise de dados em larga escala.

```python
# Leitura de arquivos no formato CSV

'''
Define a varivel "caminho" como "./", ou seja, o diretrio atual em que
o notebook est sendo executado.
'''

caminho = './'

'''
Lista os arquivos no diretrio atual por meio da funo listdir() 
do mdulo os, assim, os arquivos listados na varivel "arquivos" 
podem ser utilizados adiante partes do cdigo.
'''

arquivos = [f for f in os.listdir(caminho) if os.path.isfile(os.path.join(caminho, f)) and f.lower().endswith('.csv')]

'''
Define a funo "extrair_nome_arquivos" para extrair o nome 
dos arquivos sem a extenso ".csv" e em letras minsculas.
'''

def extrair_nome_arquivos(texto):
    nome_arquivo = texto.strip('.csv').lower()
    return nome_arquivo

'''
Lista os arquivos no diretrio atual aplicando a funo 
"extrair_nome_arquivos" em cada arquivo listado.
'''

nomes_de_arquivos = list(map(extrair_nome_arquivos, arquivos))

'''
Cria dataframes a partir dos arquivos listados usando 
a funo exec() e imprimindo uma mensagem de sucesso ou de erro.
'''

for i in range(len(nomes_de_arquivos)):
    try:
        arquivo_csv = os.path.join(caminho, arquivos[i])
        exec(f"{nomes_de_arquivos[i]} = pd.read_csv('{arquivo_csv}', sep='|', encoding='utf8')")
        print(f"Dataframe \033[34m{nomes_de_arquivos[i]}\033[0m criado a partir do arquivo \033[34m{arquivos[i]}\033[0m")
    except Exception as e:
        print("Erro ao criar o dataframe a partir do arquivo:", arquivos[i])
        print(e)
```

**2.7 Leitura de mltiplos arquivos CSV, juntando todos no mesmo dataframe**

A leitura de mltiplos arquivos CSV e sua juno em um nico dataframe  uma tarefa muito comum na anlise de dados em Python. Essa tcnica permite a anlise de grandes volumes de dados de forma organizada e eficiente. Com o uso das bibliotecas Pandas e os mtodos de concatenao e merge,  possvel juntar vrios arquivos CSV em um nico dataframe, realizando operaes de unio ou juno de acordo com as necessidades do projeto. Essa tcnica  fundamental para a anlise de dados em larga escala e pode ser facilmente implementada em Python.

```python
# Leitura de arquivos no formato CSV

'''
Define o caminho onde os arquivos CSV esto armazenados.
'''

caminho = './'

'''
Usa a biblioteca os para obter uma lista de todos 
os arquivos CSV no caminho especificado e orden-los 
em ordem alfabtica.
'''

arquivos_csv = sorted([f for f in os.listdir(caminho) if f.endswith('.csv')])

'''
Cria uma lista de dataframes, um para cada arquivo CSV 
encontrado, usando a biblioteca pandas . Os dataframes 
so lidos com o separador '|' e o encoding 'utf8'.
'''

lista_dfs = (pd.read_csv(arquivo, sep='|', encoding = 'utf8') for arquivo in arquivos_csv)

'''
Concatena todos os dataframes da lista em um nico dataframe, 
usando o mtodo concat() da biblioteca pandas. O argumento 
ignore_index=True  usado para redefinir os ndices do 
dataframe resultante.
'''

df_trienio = pd.concat(lista_dfs, ignore_index=True)
```

**2.8 Leitura de mltiplos arquivos CSV pela numerao final, com base em uma lista de anos fornecida**

A leitura de mltiplos arquivos CSV por numerao final, usando Python, proporciona uma soluo eficiente para lidar com conjuntos de dados segmentados por anos. Com base em uma lista de anos fornecida, o cdigo percorre um diretrio em busca de arquivos CSV e seleciona aqueles cuja numerao final corresponde aos anos especificados. Utilizando a biblioteca pandas, os arquivos so lidos e combinados em um nico dataframe, permitindo anlises e manipulaes integradas desses dados. Essa abordagem flexvel e automatizada simplifica o processo de leitura e consolidao de informaes de diferentes arquivos CSV, economizando tempo e esforo na anlise de dados histricos em Python.

```python
# Leitura de arquivos no formato CSV

'''
Define o caminho onde os arquivos CSV esto armazenados.
'''

caminho = './'

'''
A lista "anos" permite selecionar e processar arquivos especficos com base nos 
anos fornecidos.
'''
anos = [2009, 2010, 2011]

'''
Usa a biblioteca os para obter uma lista de todos 
os arquivos CSV no caminho especificado e orden-los 
em ordem alfabtica.
'''

arquivos_csv = [f for f in os.listdir(caminho) if f.endswith('.csv')]

'''
A lista "arquivos_selecionados"  composta por arquivos filtrados com base na 
numerao final correspondente aos anos presentes na lista "anos".
'''
arquivos_selecionados = [arquivo for arquivo in arquivos_csv if int(arquivo[-8:-4]) in anos]

'''
Cria uma lista de dataframes, um para cada arquivo CSV 
encontrado, usando a biblioteca pandas . Os dataframes 
so lidos com o separador '|' e o encoding 'utf8'.
'''

lista_dfs = [pd.read_csv(os.path.join(caminho, arquivo), sep='|', encoding='utf8') for arquivo in arquivos_selecionados]

'''
Concatena todos os dataframes da lista em um nico dataframe, 
usando o mtodo concat() da biblioteca pandas. O argumento 
ignore_index=True  usado para redefinir os ndices do 
dataframe resultante.
'''

df_trienio = pd.concat(lista_dfs, ignore_index=True)
```

Por exemplo, a data e hora atual pode ser obtida usando a funo datetime.now() e, em seguida, manipulada usando as funes do mdulo datetime. As datas so representadas como objetos datetime em Python.

## 6. Renomeando colunas

O comando rename do Python  uma funo utilizada para renomear colunas de um dataframe. Com esse comando,  possvel alterar o nome de uma ou mais colunas de um dataframe de forma rpida e simples. Alm disso,  possvel utilizar o comando para renomear ndices do dataframe. Essa funo  especialmente til quando se trabalha com dados de diferentes fontes e  necessrio padronizar os nomes das colunas para facilitar a anlise e o processamento dos dados. Com o rename,  possvel realizar essa padronizao de forma eficiente e automatizada.

```python
# Renomear manualmente, uma por uma

'''
O cdigo a seguir tem como objetivo renomear as colunas de um dataframe, sendo 
que o novo nome de cada coluna  especificado no dicionrio "colunas". A funo 
utilizada para realizar esta ao  a "rename", que recebe como argumentos o 
dicionrio com os novos nomes de cada coluna e o parmetro "inplace', que se 
verdadeiro, realiza a operao no prprio dataframe, sem criar uma cpia. Neste 
caso, a opo inplace no foi utilizada, ento a funo retorna uma cpia do 
dataframe original com as colunas renomeadas e armazenado na varivel "df". As 
colunas renomeadas foram 'NU_ANO_CENSO', 'CO_IES' e 'NO_IES', que passaram a 
se chamar 'Ano', 'Cdigo' e 'Instituio', respectivamente.
'''

colunas = {'NU_ANO_CENSO':'Ano',
           'CO_IES':'Cdigo',
           'NO_IES':'Instituio'}

df = df_trienio.rename(columns=colunas,
                      #  inplace = True
                       )

df.head()
```

```python
# Renomeando todas as colunas em letra minscula

'''
O cdigo abaixo converte as colunas do dataframe df em letras minsculas. A funo 
"str.lower"  aplicada s colunas do dataframe, que contm os nomes das colunas em 
formato de string. Isso resulta em uma nova lista de colunas com todas as letras 
em minsculo. Esse procedimento  til em casos em que as colunas foram criadas 
com letras maisculas ou misturadas e se deseja padronizar os nomes das colunas 
para facilitar a manipulao e anlise dos dados.
'''

df.columns = df.columns.str.lower()

df.head()
```

```python
# Somente a primeira letra de cada palavra em maiscula

'''
Este cdigo  usado para renomear todas as colunas do dataframe "df", transformando 
a primeira letra de cada palavra em maiscula. A funo "title"  aplicada a cada 
string do nome das colunas para executar essa transformao. Por exemplo, se uma 
coluna se chamava "nome_do_aluno", aps a execuo desse cdigo, seu nome seria 
alterado para "Nome_Do_Aluno".
'''

df.columns = df.columns.str.title()

df.head()
```

```python
# Renomeando todas as colunas em letra maiscula

'''
O cdigo abaixo converte as colunas do dataframe df em letras maisculas. A funo 
"str.lower"  aplicada s colunas do dataframe, que contm os nomes das colunas em 
formato de string. Isso resulta em uma nova lista de colunas com todas as letras 
em maisculo. Esse procedimento  til em casos em que as colunas foram criadas 
com letras minsculas ou misturadas e se deseja padronizar os nomes das colunas 
para facilitar a manipulao e anlise dos dados.
'''

df.columns = df.columns.str.upper()

df.head()
```

```python
# Remover string specficas dos nomes das colunas do dataframe

'''
O cdigo abaixo consiste em um loop "for" que percorre uma lista de substrings que 
se deseja remover do nome das colunas de um dataframe. Para cada substring na lista, 
 aplicada uma funo de substituio que remove a substring do nome das colunas, 
elimina espaos em branco extras e converte o texto resultante para letras maisculas. 
O loop percorre cada coluna do dataframe e a substituio  aplicada somente se a 
coluna  uma string e se contm a substring a ser removida. A nova lista de colunas 
resultante  armazenada de volta no dataframe. O objetivo dessa operao  limpar o 
nome das colunas, removendo prefixos e sufixos que podem atrapalhar a anlise dos dados.
'''

list_strings = ['TP_', 
                'NO_', 
                'QT_', 
                'IN_', 
                'CO_']

for str_to_replace in list_strings:
    df.columns = [col.replace(str_to_replace, '').strip().upper() if isinstance(col, str) and str_to_replace in col else col for col in df.columns]

df.head()
```

```python
# Excluso de caracteres especiais entre dos nomes de colunas

'''
O cdigo abaixo tem como objetivo substituir todos os caracteres especiais nas colunas 
do dataframe "df" por espaos em branco. Para realizar esta tarefa, o mtodo "str.replace"
 aplicado nas colunas do dataframe "df" usando a sintaxe "df.columns.str.replace()". O 
primeiro argumento da funo "replace"  o caractere que se deseja substituir, e o 
segundo argumento  o caractere que ser usado para substituir. Neste caso, o 
segundo argumento  um espao em branco (' '). Ao final do cdigo, o atributo "columns" 
do dataframe "df"  atualizado com as novas colunas que agora contm espaos em branco em 
vez de '_'.
'''

df.columns=df.columns.str.replace('_',' ')

df.head()
```

```python
# Contagem de linhas duplicadas por coluna do dataframe

'''
O cdigo define um valor para a opo "max_rows" do mdulo Pandas, que 
 usada para exibir o nmero mximo de linhas de uma tabela de dados. O valor 
atribudo  o nmero de colunas do dataframe "df_trienio" acrescido de 1, o que 
significa que todas as linhas sero exibidas na tabela.

Em seguida, o cdigo calcula o nmero de duplicatas para cada coluna do dataframe 
"df_trienio" usando o mtodo "duplicated()" e a funo "sum()" do Pandas. Essa 
contagem  armazenada na varivel "column_duplicated_counts".

Por fim, o cdigo cria um novo dataframe chamado "result_table" usando a funo 
"pd.DataFrame()". Este dataframe contm duas colunas: "Column Name" (nome da coluna) 
e "Duplicate Count" (contagem de duplicatas). Os valores dessas colunas so extrados 
da varivel "column_duplicated_counts". Por fim, a funo "display()"  usada para 
exibir a tabela "result_table" na sada. Em resumo, o cdigo mostra o nmero de 
duplicatas em cada coluna do dataframe "df_trienio".
'''

pd.options.display.max_rows = df_trienio.shape[1] + 1

column_duplicated_counts = df_trienio[df_trienio.duplicated()].sum(axis=0)

result_table = pd.DataFrame({'Nome da coluna': column_duplicated_counts.index, 'Quantidade de duplicatas': column_duplicated_counts.values})

display(result_table)
```

O filtro  uma tcnica muito utilizada na anlise de dados com Python, permitindo selecionar apenas os dados que atendem a determinadas condies. Com o uso de operadores lgicos e de comparao,  possvel selecionar linhas e colunas do dataframe que atendam a critrios especficos, facilitando a anlise de dados e a tomada de decises baseadas em informaes precisas.  uma ferramenta poderosa para lidar com grandes conjuntos de dados e facilitar a anlise exploratria.

```python
# Filtrar dados por ano

'''
O cdigo abaixo realiza a criao de um filtro para selecionar apenas as linhas do 
dataframe df_trienio que possuem o valor "2009" na coluna "NU_ANO_CENSO". Esse filtro
 criado por meio da comparao df_trienio['NU_ANO_CENSO'] == 2009. Em seguida, o 
filtro  aplicado no dataframe original utilizando o mtodo loc[], criando assim 
um novo dataframe chamado df_2009 com as linhas filtradas. Portanto, o resultado 
 um dataframe que contm apenas as informaes referentes ao ano de 2009.
'''

filtro_ano = (df_trienio['NU_ANO_CENSO'] == 2009)

df_2009 = df_trienio.loc[filtro_ano]

df_2009.head(2)
```

**11.2 Outliers**

Outliers so valores que se afastam significativamente do restante dos dados em um conjunto de dados. Eles podem ser resultado de erros de medio ou representar valores extremos dentro da populao em anlise. Identificar outliers  importante para entender a distribuio dos dados e tomar decises informadas em anlises estatsticas. Existem diversas tcnicas para identificar outliers, como a anlise de boxplot, o clculo de desvio padro e o mtodo interquartil. 

```python
# Identificao das linhas com outliers

'''
Este cdigo identifica outliers na coluna "QT_VAGAS_ANUAL_NOTURNO" do dataframe 
"df_trienio". Primeiramente, a varivel "coluna" recebe o nome da coluna que ser 
analisada. Em seguida,  calculado o primeiro quartil (Q1), o terceiro quartil 
(Q3) e o intervalo interquartil (IQR) para a coluna especificada, utilizando a 
funo "quantile" do Pandas. Com base nesses valores, so calculados o limite 
superior e o limite inferior para os dados da coluna, sendo o limite superior 
calculado atravs da soma de Q3 com 1,5 vezes o IQR, e o limite inferior calculado 
pela subtrao de Q1 com 1,5 vezes o IQR. A funo "where" do Numpy  utilizada 
para identificar as linhas do dataframe que contm valores maiores que o limite 
superior, e os ndices dessas linhas so armazenados na varivel "outlier_linhas". 
Por fim,  exibida na tela uma mensagem contendo a lista de ndices das linhas do 
dataframe que contm outliers para a coluna especificada. Esse tipo de cdigo  
til para identificar e tratar outliers em colunas especficas, garantindo uma 
anlise mais precisa dos dados.
'''

coluna = 'QT_VAGAS_ANUAL_NOTURNO'

Q1 = df_trienio[coluna].quantile(0.25)

Q3 = df_trienio[coluna].quantile(0.75)

IQR = Q3 - Q1

limite_superior = Q3 + (1.5 * IQR)
                      
limite_inferior = Q1 - (1.5 * IQR)                        

outlier_linhas = np.where(df_trienio[coluna] > limite_superior)[0]

print('Linhas com outliers para a', coluna, 'do dataframe df_trienio: \n\n', outlier_linhas)
```

```python
# Nmero de outliers por coluna que inicia com "QT_"

'''
Este cdigo conta o nmero de outliers em cada coluna que comea com "QT_" do 
dataframe "df_trienio".

Primeiro, as colunas que comeam com "QT_" so selecionadas usando o mtodo "filter" 
e armazenadas na varivel "df". Em seguida,  definida uma funo chamada "contar_outliers" 
que recebe uma coluna como entrada. Dentro dessa funo, a coluna  filtrada para 
remover os valores igual a zero (zeros podem ser usados para preencher campos vazios). 
Em seguida,  calculado o primeiro e terceiro quartil (Q1 e Q3) da coluna e, a partir 
disso,  calculado o intervalo interquartil (IQR) e o limiar para identificar outliers 
(limiar  definido como 1,5 vezes o IQR). Por fim,  retornado o nmero de valores 
que so identificados como outliers.

Na sequncia, a funo "contar_outliers"  aplicada a todas as colunas de "df" usando 
o mtodo "apply" e o resultado  armazenado na varivel "outliers_por_coluna". Por fim, 
 feita uma impresso com o nmero de outliers encontrados por coluna.
'''

df = df_trienio.filter(regex='^QT_')

def contar_outliers(coluna):
    coluna = coluna[coluna != 0] 
    if len(coluna) == 0: 
        return 0
    q1, q3 = np.percentile(coluna, [25, 75])
    iqr = q3 - q1
    limiar = 1.5 
    
    outliers = (coluna < q1 - limiar*iqr) | (coluna > q3 + limiar*iqr)
    return np.sum(outliers)

outliers_por_coluna = df.apply(contar_outliers)

outliers_por_coluna = outliers_por_coluna[outliers_por_coluna != 0]

print('Nmero de outliers por coluna:\n\n', outliers_por_coluna)
```

```python
# Intervalo Interquartil

'''
Este cdigo calcula o intervalo interquartil (IQR) para todas as colunas do DataFrame 
"df_trienio" que iniciam com os caracteres "QT_", utilizando as funes "filter" e 
"quantile" do Pandas. Primeiramente, a varivel "Q1"  criada para armazenar o primeiro 
quartil (25%) dos dados de todas as colunas que atendem ao filtro. Em seguida, a 
varivel "Q3"  criada para armazenar o terceiro quartil (75%) dos mesmos dados. 
Com base nesses valores,  possvel calcular o IQR para todas as colunas utilizando 
a subtrao de Q1 de Q3. A varivel "non_zero_IQR"  criada para armazenar apenas os 
valores de IQR que so diferentes de zero, filtrando as colunas que no possuem 
variao nos dados. Por fim,  exibido na tela o valor de "non_zero_IQR", que contm 
uma lista dos valores de IQR para as colunas que possuem variao nos dados. Esse 
tipo de cdigo  til para identificar as colunas que possuem uma variao significativa 
nos dados, permitindo uma anlise mais precisa e focada nessas colunas especficas.
'''

Q1 = df_trienio.filter(regex='^QT_').quantile(0.25)

Q3 = df_trienio.filter(regex='^QT_').quantile(0.75)

IQR = Q3 - Q1

nao_zero_IQR = IQR[IQR != 0]

print(nao_zero_IQR)
```

```python
# Remoo de outliers - Apenas graficamente

'''
Este cdigo plota um boxplot da varivel "QT_VAGAS_ANUAL_NOTURNO" do DataFrame 
"df_trienio" usando a biblioteca Seaborn. O parmetro "x" especifica que a varivel 
"QT_VAGAS_ANUAL_NOTURNO" deve ser usada no eixo horizontal do grfico. O parmetro 
"showfliers"  definido como "False", o que significa que os outliers sero excludos 
do boxplot. O comando "plt.xlabel" adiciona um rtulo ao eixo horizontal do grfico 
e "plt.show" exibe o grfico.
'''

sns.boxplot(x=df_trienio['QT_VAGAS_ANUAL_NOTURNO'], showfliers=False)

plt.xlabel('\n Vagas Anual Noturno')

plt.show()
```

```python
# Remoo de outliers - No dataframe

'''
Este cdigo filtra o dataframe df_trienio para incluir apenas as colunas 
que comeam com 'QT_'. Em seguida, calcula o primeiro quartil (Q1), o terceiro 
quartil (Q3) e a amplitude interquartil (IQR) para cada uma dessas colunas. O 
limite inferior (lim_inferior) e superior (lim_superior) para identificar os outliers 
so calculados com base no IQR. Depois, um novo dataframe  criado (df_sem_outliers)
com todas as linhas que no contm nenhum outlier para as colunas selecionadas. 
Este novo dataframe  retornado como resultado. A funo "all(axis=1)"  usada para 
garantir que as linhas com outliers em qualquer coluna sejam removidas.
'''

df = df_trienio.filter(regex='^QT_')

Q1 = df.quantile(0.25)

Q3 = df.quantile(0.75)

IQR = Q3 - Q1
 
lim_inferior = Q1 - 1.5 * IQR

lim_superior = Q3 + 1.5 * IQR

df_sem_outliers = df.loc[((df >= lim_inferior) & (df <= lim_superior)).all(axis=1)]

df_sem_outliers.head()
```

```python
# Diagrama de caixas aps a remoo dos outliers no dataframe

'''
Este cdigo gera um boxplot da coluna "QT_VAGAS_ANUAL_NOTURNO" do dataframe 
"df_sem_outliers", utilizando a biblioteca Seaborn.
'''

sns.boxplot(x=df_sem_outliers['QT_VAGAS_ANUAL_NOTURNO'])

plt.xlabel('\n Vagas Anual Noturno')

plt.show()
```

Grficos de barras so representaes visuais de dados em que informaes so apresentadas por meio de barras retangulares. ### Barras empilhadas

Grficos de barras empilhadas so uma variao dos grficos de barras tradicionais, que permitem visualizar a composio dos dados de uma maneira mais detalhada. Em um grfico de barras empilhadas, cada barra  dividida em segmentos coloridos que representam diferentes categorias ou subcategorias de dados. Esses segmentos so empilhados uns sobre os outros, permitindo que se visualize a contribuio de cada categoria para o total representado pela barra. Essa tcnica  til para comparar a distribuio de categorias em diferentes grupos de dados.

```python
# Grfico de barras empilhado

'''
O cdigo a seguir realiza a plotagem de um grfico de barras para exibir os totais 
de concluintes e ingressantes por ano. Primeiramente,  feito um agrupamento dos 
dados pelo ano censitrio e calculada a soma dos valores de concluintes e ingressantes 
para cada ano. Em seguida,  determinado o valor mximo para o eixo y, considerando 
os totais de ingressantes e concluintes, multiplicado por 1.1 para adicionar uma 
margem. O ndice do DataFrame  convertido para uma representao de string. A 
partir disso,  criada uma figura e um conjunto de eixos. A varivel x  criada 
para representar a posio das barras no eixo x. As barras de ingressantes e 
concluintes so desenhadas usando os dados correspondentes, e so atribudas as 
cores laranja e azul, respectivamente. O eixo x  rotulado como "Ano" e as legendas 
so adicionadas. O limite do eixo y  definido com base no valor mximo calculado 
anteriormente. O formato dos rtulos do eixo y  definido como um nmero inteiro. 
Os rtulos do eixo x so definidos com base nos valores do ndice do DataFrame. 
Finalmente, o grfico  exibido.
'''

soma_por_ano = df_trienio.groupby('NU_ANO_CENSO')[['QT_CONCLUINTE_TOTAL', 'QT_INGRESSO_TOTAL']].sum()
max_y = max(soma_por_ano[['QT_INGRESSO_TOTAL', 'QT_CONCLUINTE_TOTAL']].values.flatten()) * 1.1

soma_por_ano.index = soma_por_ano.index.astype(str)  

fig, ax = plt.subplots()
x = np.arange(len(soma_por_ano.index))  
ax.bar(x, soma_por_ano['QT_INGRESSO_TOTAL'], label='Ingressantes', color='orange')
ax.bar(x, soma_por_ano['QT_CONCLUINTE_TOTAL'], label='Concluintes', color='blue')
ax.set_xlabel('Ano')
# ax.set_ylabel('Total')
# ax.set_title('Total de concluintes e ingressantes por ano')

ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2)

ax.set_ylim(0, max_y)
ax.ticklabel_format(axis='y', style='plain')
ax.set_xticks(x)  
ax.set_xticklabels(soma_por_ano.index) 

plt.show()
```

### Mapas de calor

Grficos do tipo mapa de calor so uma representao visual de dados que utiliza cores para indicar a magnitude de uma varivel em relao a outras. Geralmente, so utilizados em anlises de dados multidimensionais, em que  necessrio identificar padres ou correlaes entre variveis. O mapa de calor  construdo a partir de uma matriz de dados, em que as linhas representam uma varivel e as colunas representam outra. Cada clula da matriz  preenchida com uma cor que representa o valor da varivel em relao s outras. Quanto mais escuro o tom da cor, maior o valor da varivel, e quanto mais claro, menor o valor.

```python
# Mapa de calor

'''
Este cdigo cria um heatmap da correlao entre algumas variveis do dataframe 
df_trienio, incluindo QT_MATRICULA_TOTAL, QT_CONCLUINTE_TOTAL, QT_INGRESSO_PROCESSO_SELETIVO, 
QT_VAGA_TOTAL e QT_INSCRITO_TOTAL. Para isso,  utilizada a funo corr() do pandas 
para calcular a correlao entre essas variveis e em seguida, a biblioteca Seaborn 
 utilizada para plotar o heatmap. A intensidade da cor indica a fora e direo 
da correlao, enquanto que a escala de cores azul e vermelho indica a direo da 
correlao (negativa ou positiva, respectivamente). A opo annot=True adiciona os 
valores de correlao dentro de cada clula do heatmap.
'''

plt.figure(figsize=(10,6))

corr = df_trienio[['QT_MATRICULA_TOTAL', 
                   'QT_CONCLUINTE_TOTAL', 
                   'QT_INGRESSO_PROCESSO_SELETIVO', 
                   'QT_VAGA_TOTAL', 
                   'QT_INSCRITO_TOTAL']].corr()

sns.heatmap(corr,
            cmap='coolwarm',
            annot=True
            );

plt.show()
```

### Histograma

Um histograma  um grfico de frequncia que representa a distribuio de valores de uma varivel quantitativa. Ele  construdo a partir da diviso do intervalo dos valores da varivel em vrias faixas, chamadas de "bins", e contabiliza quantos valores da varivel esto dentro de cada bin. Assim, o eixo x do histograma representa os valores da varivel, enquanto o eixo y representa a contagem de observaes dentro de cada bin. O histograma  til para visualizar a forma da distribuio dos dados, bem como a presena de eventuais assimetrias, outliers ou outras caractersticas relevantes.

```python
# Histograma do total de ingressantes por IES

'''
Este cdigo cria um histograma que representa a distribuio da quantidade total 
de ingressantes por instituio de ensino superior (IES) em um perodo de trs 
anos. Primeiramente,  feito um agrupamento dos dados do dataframe "df_trienio" 
pela coluna "CO_IES" (cdigo da IES) e a soma da quantidade de ingressantes de cada 
uma delas  obtida. Em seguida,  definido o nmero de "bins" (intervalos) para 
o histograma e a figura  criada com o tamanho de 10x6. O histograma  gerado 
utilizando a funo "hist" do Matplotlib, que recebe como parmetro os valores 
do agrupamento, o nmero de intervalos e a cor laranja. So definidos os rtulos 
dos eixos x e y, limites dos eixos x e y, bem como a escala da marcao do eixo x. 
So calculados os percentuais de ocorrncia de cada intervalo e  verificado se 
algum intervalo possui mais de 2% de ocorrncia. Caso haja,  adicionado ao grfico 
um texto com a porcentagem. Tambm  desenhada uma linha vermelha que representa 
a mediana da quantidade de ingressantes por IES.
'''

grupo = df_trienio.groupby('CO_IES')['QT_INGRESSO_TOTAL'].sum()

bins = 100

fig, ax = plt.subplots(figsize=(10, 6))

hist, edges, _ = ax.hist(grupo.values, bins=bins, color='orange')

ax.set_xlabel('\nIngressantes por ano no trinio 2009-2011')
ax.set_ylabel('Quantidade de IESs\n')

ax.set_xlim([0, df_trienio['QT_INGRESSO_TOTAL'].max()])
ax.set_xticks([i*1000 for i in range(int(df_trienio['QT_INGRESSO_TOTAL'].max()/1000)+1)]) 
ax.ticklabel_format(style='sci', axis='x', scilimits=(3,3), useMathText=True)

ylim = max(hist)
ax.set_ylim([0, ylim + 0.15 * ylim])

total = len(grupo)
percentages = (hist / total) * 100
max_index = np.argmax(hist)

if percentages[max_index] >= 2:
    ax.text(edges[max_index]+(edges[max_index+1]-edges[max_index])/2,\
            hist[max_index], f'{percentages[max_index]:.1f}%', ha='left',\
            va='bottom', color='green')

median = np.median(grupo.values)
ax.axvline(median, color='red')
ax.text(median*1.05, ylim*1.1, ' Mediana', color='red')

plt.show()
```

```python
# Histograma do total de concluintes por IES

'''
Este cdigo cria um histograma que representa a distribuio da quantidade total 
de concluintes por instituio de ensino superior (IES) em um perodo de trs 
anos. Primeiramente,  feito um agrupamento dos dados do dataframe "df_trienio" 
pela coluna "CO_IES" (cdigo da IES) e a soma da quantidade de concluintes de cada 
uma delas  obtida. Em seguida,  definido o nmero de "bins" (intervalos) para 
o histograma e a figura  criada com o tamanho de 10x6. O histograma  gerado 
utilizando a funo "hist" do Matplotlib, que recebe como parmetro os valores 
do agrupamento, o nmero de intervalos e a cor laranja. So definidos os rtulos 
dos eixos x e y, limites dos eixos x e y, bem como a escala da marcao do eixo x. 
So calculados os percentuais de ocorrncia de cada intervalo e  verificado se 
algum intervalo possui mais de 2% de ocorrncia. Caso haja,  adicionado ao grfico 
um texto com a porcentagem. Tambm  desenhada uma linha vermelha que representa 
a mediana da quantidade de concluintes por IES.
'''

grupo = df_trienio.groupby('CO_IES')['QT_CONCLUINTE_TOTAL'].sum()

bins = 100

fig, ax = plt.subplots(figsize=(10, 6))

hist, edges, _ = ax.hist(grupo.values, bins=bins, color='blue')

ax.set_xlabel('\nConcluintes no trinio 2009-2011')
ax.set_ylabel('Quantidade de IESs\n')

ax.set_xlim([0, df_trienio['QT_CONCLUINTE_TOTAL'].max()])
ax.set_xticks([i*1000 for i in range(int(df_trienio['QT_CONCLUINTE_TOTAL'].max()/1000)+1)]) 
ax.ticklabel_format(style='sci', axis='x', scilimits=(3,3), useMathText=True)

ylim = max(hist)
ax.set_ylim([0, ylim + 0.25 * ylim])

total = len(grupo)
percentages = (hist / total) * 100
max_index = np.argmax(hist)

if percentages[max_index] >= 2:
    ax.text(edges[max_index]+(edges[max_index+1]-edges[max_index])/2, hist[max_index],\
            f'{percentages[max_index]:.1f}%', ha='left', va='bottom', color='green')

median = np.median(grupo.values)
ax.axvline(median, color='red')
ax.text(median*1.05, ylim*1.15, ' Mediana', color='red')

plt.show()
```

No Google Colab,  possvel fazer o download de arquivos gerados durante a execuo do cdigo diretamente para o computador pessoal do usurio. Para isso, basta usar o mdulo files da biblioteca google.colab e o mtodo download() passando como parmetro o caminho do arquivo desejado no Google Drive. 

```python
# Dataframes - um a um

'''
O cdigo abaixo usa a biblioteca Google Colab para baixar um arquivo chamado 
"trienio_2009_2010_2011.csv" localizado no diretrio atual especificado pelo caminho 
"./". O comando "gc.files.download"  usado para iniciar o download do arquivo para 
a mquina local do usurio. Ao executar o comando, o Google Colab inicia o processo 
de download do arquivo especificado pelo caminho e envia-o para o navegador do usurio. 
Depois disso, o usurio pode optar por salvar o arquivo em qualquer diretrio local 
do seu computador.  importante notar que o comando "gc.files.download"  especfico 
para o ambiente do Google Colab e no pode ser executado em outros ambientes de 
programao.
'''

caminho = './'

gc.files.download(caminho + 'trienio_2009_2010_2011.csv')
```

```python
# Dataframes - arquivo compactado

'''
O cdigo a seguir tem como objetivo compactar o arquivo CSV "trienio_2009_2010_2011.csv" 
em um arquivo ZIP chamado "trienio_2009_2010_2011.zip". O caminho do diretrio 
onde o arquivo est localizado  definido como "./".
Usando a biblioteca zipfile, o cdigo abre um arquivo ZIP em modo de escrita e, 
em seguida, escreve o arquivo CSV dentro do arquivo ZIP com o mesmo nome. O 
mtodo write()  utilizado para adicionar o arquivo CSV ao arquivo ZIP.
Por fim,  feito o download do arquivo ZIP compactado usando a funo 
files.download() do Google Colab, permitindo que o usurio faa o download do 
arquivo resultante. Isso torna mais fcil para o usurio obter o arquivo 
compactado contendo o arquivo CSV desejado.
'''

caminho = './'
arquivo_csv = 'trienio_2009_2010_2011.csv'
arquivo_zip = 'trienio_2009_2010_2011.zip'


with zipfile.ZipFile(arquivo_zip, 'w') as zipf:
    zipf.write(os.path.join(caminho, arquivo_csv), arquivo_csv)

gc.files.download(arquivo_zip)
```

```python
# Dataframes - todos os arquivos de um diretrio

'''
O cdigo a seguir usa a biblioteca glob para buscar todos os arquivos CSV no diretrio 
atual especificado pelo caminho "./". O comando "glob.glob"  usado para buscar todos
os arquivos que correspondem ao padro especificado no caminho, que neste caso  "*.csv", 
ou seja, todos os arquivos com extenso CSV. A biblioteca "os.path"  usada para unir 
o caminho e o padro de busca em um nico caminho. Em seguida, o cdigo usa um loop "for" 
para baixar cada um dos arquivos encontrados. O comando "gc.files.download"  usado 
para baixar cada arquivo encontrado pelo loop "for". Dessa forma, o cdigo permite 
que o usurio baixe todos os arquivos CSV em um diretrio especfico.  importante 
notar que o comando "gc.files.download"  especfico para o ambiente do Google Colab e 
no pode ser executado em outros ambientes de programao.
'''

caminho = './'

csv_arquivo = glob.glob(os.path.join(caminho, '*.csv'))

for arquivo in csv_arquivo:
    gc.files.download(arquivo)
```

```python
# Dataframes - todos os arquivos de um diretrio compactados

'''
O cdigo abaixo tem como objetivo compactar os arquivos CSV presentes no diretrio 
especificado em um arquivo ZIP. Primeiro,  utilizado o mdulo glob para obter 
uma lista de todos os arquivos com extenso .csv no diretrio. Em seguida, 
definido o nome do arquivo ZIP como "arquivos_csv.zip".
Dentro do bloco with, um arquivo ZIP  criado utilizando a biblioteca zipfile. 
Em um loop, cada arquivo CSV  adicionado ao arquivo ZIP utilizando o mtodo write() 
do objeto zipf. O nome de cada arquivo dentro do arquivo ZIP  definido utilizando 
a funo os.path.basename(), garantindo que apenas o nome do arquivo seja utilizado.
Por fim,  feito o download do arquivo ZIP compactado utilizando a funo 
files.download() do Google Colab, permitindo que o usurio faa o download do 
arquivo resultante contendo os arquivos CSV compactados em um nico arquivo ZIP. 
Isso facilita o compartilhamento e transporte dos arquivos CSV em uma nica 
unidade compactada.
'''

caminho = './'
arquivos_csv = glob.glob(os.path.join(caminho, '*.csv'))
arquivo_zip = 'arquivos_compactados.zip'

with zipfile.ZipFile(arquivo_zip, 'w') as zipf:
   
    for arquivo in arquivos_csv:
        zipf.write(arquivo, os.path.basename(arquivo))


gc.files.download(arquivo_zip)
```

```python
# Dataframes - apenas alguns arquivos de um diretrio compactados

'''
O cdigo a seguir tem como objetivo compactar apenas os arquivos indicados na lista 
arquivos em um arquivo ZIP chamado "arquivos_compactados.zip".
Dentro do bloco with,  criado um arquivo ZIP utilizando a biblioteca zipfile. 
Em um loop, cada arquivo da lista arquivos  verificado se existe no diretrio 
especificado usando a funo os.path.isfile(). Caso o arquivo exista, ele  
adicionado ao arquivo ZIP utilizando o mtodo write() do objeto zipf, com o 
nome do arquivo mantido.
Por fim,  feito o download do arquivo ZIP compactado utilizando a funo 
files.download() do Google Colab, permitindo que o usurio faa o download 
do arquivo resultante contendo apenas os arquivos indicados na lista 
pr-fornecida. Isso facilita o compartilhamento e transporte especfico 
desses arquivos em um nico arquivo compactado.
'''

caminho = './'

arquivos = [
            'SoU_censo_IES_2009.csv', 
            'SoU_censo_IES_2011.csv', 
            'SoU_censo_IES_2010.csv', 
            'trienio_2009_2010_2011.csv'
           ]

arquivo_zip = 'arquivos_compactados.zip'

with zipfile.ZipFile(arquivo_zip, 'w') as zipf:
    for arquivo in arquivos:
        caminho_arquivo = os.path.join(caminho, arquivo)
        if os.path.isfile(caminho_arquivo):
            zipf.write(caminho_arquivo, arquivo)

gc.files.download(arquivo_zip)
```
