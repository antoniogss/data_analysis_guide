## Introdução

**O que é Análise Exploratória de Dados?**

Análise Exploratória de Dados (AED) é um processo de investigação utilizado para entender e analisar conjuntos de dados de forma sistemática e detalhada. O objetivo principal da AED é extrair *insights* e conhecimentos relevantes sobre os dados, identificando padrões, tendências e anomalias. É uma etapa fundamental em qualquer projeto de análise de dados, pois ajuda a entender a qualidade e a integridade dos dados e pode levar a descobertas que orientam a tomada de decisões embasadas.


**Como realizar uma Análise Exploratória de Dados?**

Para realizar uma AED, é importante seguir alguns passos básicos. Primeiramente, é necessário ter uma compreensão clara dos objetivos da análise e do conjunto de dados em questão. Em seguida, é recomendado explorar as estatísticas descritivas básicas, como média, mediana e desvio padrão, e visualizações gráficas, como diagramas de caixas e gráficos de dispersão. Além disso, é importante investigar a distribuição dos dados e identificar possíveis valores extremos e valores ausentes. Por fim, a AED deve ser iterativa, ou seja, é preciso voltar a alguns passos e refinar a análise à medida que novas informações são descobertas.

```python
# Identificar o separador ("sep") dos arquivos CSV

# O código Shell Script abaixo tem como objetivo identificar o separador utilizado 
# em arquivos CSV de um diretório específico. Ele percorre cada arquivo CSV no 
# diretório e realiza testes para encontrar o separador, usando uma lista de 
# possíveis delimitadores.  Ao encontrar um separador no arquivo, ele exibe essa 
# informação. Essa abordagem é útil para entender a estrutura dos arquivos CSV e 
# facilitar o processamento dos dados posteriormente.

%%shell

caminho="./"

for arquivo_csv in $caminho*.csv; do
    echo "Arquivo: $arquivo_csv"

    delimitadores=("," ";" "\t" "|")

    for delimitador in "${delimitadores[@]}"; do
        resultado=$(head -n 1 "$arquivo_csv" | grep -o "$delimitador" | wc -l)
        
        if [ "$resultado" -gt 0 ]; then
            echo "Separador: $delimitador"
            break
        fi
    done

    echo "-------------------------------------"
done
```

**2.6 Leitura de múltiplos arquivos CSV em dataframes individuais, através de um loop (for)**

Uma das tarefas mais comuns na análise de dados é a leitura de arquivos CSV e sua transformação em dataframes do Pandas. Nesse sentido, a leitura de múltiplos arquivos CSV em dataframes individuais através de um loop (for) é uma técnica muito útil e eficiente em Python. Com essa técnica, é possível ler e manipular grandes quantidades de dados de forma organizada e automatizada, o que é fundamental para a análise de dados em larga escala.

```python
# Leitura de arquivos no formato CSV

'''
Define a variável "caminho" como "./", ou seja, o diretório atual em que
o notebook está sendo executado.
'''

caminho = './'

'''
Lista os arquivos no diretório atual por meio da função listdir() 
do módulo os, assim, os arquivos listados na variável "arquivos" 
podem ser utilizados adiante partes do código.
'''

arquivos = [f for f in os.listdir(caminho) if os.path.isfile(os.path.join(caminho, f)) and f.lower().endswith('.csv')]

'''
Define a função "extrair_nome_arquivos" para extrair o nome 
dos arquivos sem a extensão ".csv" e em letras minúsculas.
'''

def extrair_nome_arquivos(texto):
    nome_arquivo = texto.strip('.csv').lower()
    return nome_arquivo

'''
Lista os arquivos no diretório atual aplicando a função 
"extrair_nome_arquivos" em cada arquivo listado.
'''

nomes_de_arquivos = list(map(extrair_nome_arquivos, arquivos))

'''
Cria dataframes a partir dos arquivos listados usando 
a função exec() e imprimindo uma mensagem de sucesso ou de erro.
'''

for i in range(len(nomes_de_arquivos)):
    try:
        arquivo_csv = os.path.join(caminho, arquivos[i])
        exec(f"{nomes_de_arquivos[i]} = pd.read_csv('{arquivo_csv}', sep='|', encoding='utf8')")
        print(f"Dataframe \033[34m{nomes_de_arquivos[i]}\033[0m criado a partir do arquivo \033[34m{arquivos[i]}\033[0m")
    except Exception as e:
        print("Erro ao criar o dataframe a partir do arquivo:", arquivos[i])
        print(e)
```

**2.7 Leitura de múltiplos arquivos CSV, juntando todos no mesmo dataframe**

A leitura de múltiplos arquivos CSV e sua junção em um único dataframe é uma tarefa muito comum na análise de dados em Python. Essa técnica permite a análise de grandes volumes de dados de forma organizada e eficiente. Com o uso das bibliotecas Pandas e os métodos de concatenação e merge, é possível juntar vários arquivos CSV em um único dataframe, realizando operações de união ou junção de acordo com as necessidades do projeto. Essa técnica é fundamental para a análise de dados em larga escala e pode ser facilmente implementada em Python.

```python
# Leitura de arquivos no formato CSV

'''
Define o caminho onde os arquivos CSV estão armazenados.
'''

caminho = './'

'''
Usa a biblioteca os para obter uma lista de todos 
os arquivos CSV no caminho especificado e ordená-los 
em ordem alfabética.
'''

arquivos_csv = sorted([f for f in os.listdir(caminho) if f.endswith('.csv')])

'''
Cria uma lista de dataframes, um para cada arquivo CSV 
encontrado, usando a biblioteca pandas . Os dataframes 
são lidos com o separador '|' e o encoding 'utf8'.
'''

lista_dfs = (pd.read_csv(arquivo, sep='|', encoding = 'utf8') for arquivo in arquivos_csv)

'''
Concatena todos os dataframes da lista em um único dataframe, 
usando o método concat() da biblioteca pandas. O argumento 
ignore_index=True é usado para redefinir os índices do 
dataframe resultante.
'''

df_trienio = pd.concat(lista_dfs, ignore_index=True)
```

**2.8 Leitura de múltiplos arquivos CSV pela numeração final, com base em uma lista de anos fornecida**

A leitura de múltiplos arquivos CSV por numeração final, usando Python, proporciona uma solução eficiente para lidar com conjuntos de dados segmentados por anos. Com base em uma lista de anos fornecida, o código percorre um diretório em busca de arquivos CSV e seleciona aqueles cuja numeração final corresponde aos anos especificados. Utilizando a biblioteca pandas, os arquivos são lidos e combinados em um único dataframe, permitindo análises e manipulações integradas desses dados. Essa abordagem flexível e automatizada simplifica o processo de leitura e consolidação de informações de diferentes arquivos CSV, economizando tempo e esforço na análise de dados históricos em Python.

```python
# Leitura de arquivos no formato CSV

'''
Define o caminho onde os arquivos CSV estão armazenados.
'''

caminho = './'

'''
A lista "anos" permite selecionar e processar arquivos específicos com base nos 
anos fornecidos.
'''
anos = [2009, 2010, 2011]

'''
Usa a biblioteca os para obter uma lista de todos 
os arquivos CSV no caminho especificado e ordená-los 
em ordem alfabética.
'''

arquivos_csv = [f for f in os.listdir(caminho) if f.endswith('.csv')]

'''
A lista "arquivos_selecionados" é composta por arquivos filtrados com base na 
numeração final correspondente aos anos presentes na lista "anos".
'''
arquivos_selecionados = [arquivo for arquivo in arquivos_csv if int(arquivo[-8:-4]) in anos]

'''
Cria uma lista de dataframes, um para cada arquivo CSV 
encontrado, usando a biblioteca pandas . Os dataframes 
são lidos com o separador '|' e o encoding 'utf8'.
'''

lista_dfs = [pd.read_csv(os.path.join(caminho, arquivo), sep='|', encoding='utf8') for arquivo in arquivos_selecionados]

'''
Concatena todos os dataframes da lista em um único dataframe, 
usando o método concat() da biblioteca pandas. O argumento 
ignore_index=True é usado para redefinir os índices do 
dataframe resultante.
'''

df_trienio = pd.concat(lista_dfs, ignore_index=True)
```

## 4. Tipos de dados em Python

**4.1 Identificando os tipos de dados**

1.   **Inteiros** (*int64*) são um tipo de dado numérico em Python que representa números inteiros sem decimais. Por exemplo, 1, 10, -5 são inteiros.
2.   **Floats** (*float*) são um tipo de dado numérico em Python que representa números com decimais. Por exemplo, 1.5, -0.75, 3.1415 são floats.
3.   **Strings** (*object*) são um tipo de dado em Python que representa texto ou sequências de caracteres. As strings são definidas entre aspas simples ou duplas. Por exemplo, 'Olá', "Mundo", "Python é uma linguagem de programação" são strings.
4.   **Datas** (*datetime64*) são um tipo de dado em Python que representa datas e horas. O módulo datetime do Python permite trabalhar com datas, horas e operações de tempo. Por exemplo, a data e hora atual pode ser obtida usando a função datetime.now() e, em seguida, manipulada usando as funções do módulo datetime. As datas são representadas como objetos datetime em Python.

Ao trabalhar com dataframes em Python, é comum precisar selecionar apenas algumas colunas específicas para realizar uma análise ou para criar um novo dataframe.
```python
# Selecionar apenas as colunas numéricas (int64 e float)

'''
O código abaixo seleciona apenas as colunas numéricas do dataframe "df_trienio" 
e armazena em um novo dataframe chamado "df_numerico". A função "select_dtypes" 
é utilizada para filtrar as colunas do dataframe por tipo de dados, mas neste caso 
é utilizado o parâmetro exclude, que indica que as colunas cujo tipo de dados é 
especificado devem ser excluídas da seleção. No caso, o parâmetro passado foi 
'object', indicando que as colunas do tipo objeto devem ser excluídas da seleção. 
Dessa forma, apenas as colunas numéricas (inteiros e floats) são selecionadas e 
armazenadas no novo dataframe "df_numerico". Esse tipo de seleção pode ser útil 
em análises em que apenas as colunas numéricas são relevantes ou para transformação 
e manipulação dos dados numéricos do conjunto de dados.
'''

df_numerico = df_trienio.select_dtypes(exclude='object')

df_numerico.head()
```

```python
# Selecionar colunas cujo nome possui uma string específica

'''
O código abaixo seleciona apenas as colunas do dataframe "df_trienio" cujos nomes 
começam com a string "QT" e armazena em um novo dataframe chamado "df_QT". A função 
"filter" é utilizada para filtrar as colunas do dataframe pelos seus nomes, utilizando 
expressões regulares. No caso, a expressão regular "^QT" indica que serão selecionadas 
apenas as colunas cujo nome começa com as letras "QT" e armazenadas no novo dataframe 
"df_QT". Essa seleção pode ser útil em análises onde apenas as colunas que apresentam 
uma característica comum nos nomes são relevantes ou para seleção de um subconjunto 
específico de colunas do conjunto de dados.
'''

df_QT = df_trienio.filter(regex = '^QT.*', axis=1)

df_QT.head()
```

```python
# Selecionar colunas cujo nome possui uma string específica em qualquer posição

'''
O código acima seleciona as colunas do dataframe ¨df_trienio" cujos nomes contêm 
a string 'INSC_VAGA' em qualquer posição do nome. A função "filter" é utilizada 
para filtrar as colunas do dataframe pelos seus nomes, utilizando expressões 
regulares. No caso, a expressão regular 'INSC_VAGA' indica que serão selecionadas 
apenas as colunas cujo nome contém a string 'INSC_VAGA' e armazenadas no novo 
dataframe "df_VAGA". Essa seleção pode ser útil em análises onde apenas as 
colunas que apresentam uma característica comum nos nomes são relevantes ou 
para seleção de um subconjunto específico de colunas do conjunto de dados.
'''

df_INSC_VAGA = df_trienio.filter(regex = 'INSC_VAGA', axis=1)

df_INSC_VAGA.head()
```

```python
# Selecionar colunas cujo nome termina com uma string específica

'''
O código abaixo seleciona as colunas do dataframe "df_trienio" cujos nomes 
terminam com a string 'TOTAL'. A função "filter" é utilizada para filtrar as 
colunas do dataframe pelos seus nomes, utilizando expressões regulares. 
No caso, a expressão regular '.*TOTAL$' indica que serão selecionadas apenas as 
colunas cujo nome termina com a string 'TOTAL'e armazenadas no novo dataframe 
"df_TOTAL". Essa seleção pode ser útil em análises onde apenas as colunas 
que apresentam um total são relevantes ou para seleção de um subconjunto 
específico de colunas do conjunto de dados.
'''

df_TOTAL = df_trienio.filter(regex='TOTAL$', axis=1)

df_TOTAL.head()
```

```python
# Selecionar colunas cujo nome possui duas (ou mais) strings específicas

'''
Este código utiliza a função filter para selecionar apenas as colunas do dataframe 
"df_trienio" que contêm a string 'QT' em seu nome, utilizando o argumento "regex" 
com a expressão regular 'QT' para encontrar essa correspondência. Em seguida, a 
função filter é aplicada novamente para selecionar apenas as colunas desse subconjunto 
que contêm a string 'VAGA' em seu nome, utilizando a expressão regular 'VAGA'. O 
resultado é um novo dataframe, "df_QT_e_VAGA", que contém apenas as colunas que 
correspondem a ambas as expressões regulares, ou seja, as colunas que contêm 
'QT' no início do nome e 'VAGA' em qualquer posição do nome.
'''

df_QT_e_VAGA = df_trienio.filter(regex='^QT', axis=1).filter(regex='VAGA', axis=1)

df_QT_e_VAGA.head()
```

```python
# Selecionar colunas cujo nome possui uma ou outra strings específicas, ou as duas.

'''
Este código seleciona as colunas do dataframe "df_trienio" que possuem a string 
'QT' ou 'VAGA' em qualquer posição do nome da coluna. A função "filter" é utilizada 
para filtrar as colunas, e o argumento "regex" é usado para especificar os padrões 
a serem filtrados. O operador "|" dentro do argumento "regex" indica que a seleção 
deve incluir colunas que atendam a qualquer um dos padrões especificados. O resultado 
é armazenado no dataframe "df_QT_ou_VAGA".
'''

df_QT_ou_VAGA = df_trienio.filter(regex='QT|VAGA', axis=1)

df_QT_ou_VAGA.head()
```

**5.5 Alterando a coluna índice do dataframe**

O comando "set_index" do Python é utilizado para definir uma coluna específica do dataframe como índice. Com isso, é possível acessar as informações de cada linha através do seu índice, o que pode facilitar a manipulação dos dados em determinadas situações. Além disso, o "set_index" pode ser usado para remover uma coluna do dataframe e torná-la índice ao mesmo tempo, permitindo que a informação seja mantida, mas não fique duplicada no conjunto de dados. É uma função útil para organizar e explorar dados, principalmente em análises que exigem um controle mais preciso dos índices das informações.

```python
# Selecionando uma coluna específica do dataframe para ser o índice

'''
Esse código define a coluna "NU_ANO_CENSO" como índice do dataframe "df_trienio" 
usando o método "set_index". Ao definir uma coluna como índice, todas as operações 
subsequentes no dataframe serão realizadas com base nessa nova estrutura de índice. 
Se o argumento "inplace=True" for usado, o dataframe original será modificado e o 
resultado será salvo diretamente nele, caso contrário, será retornado um novo dataframe 
com o índice modificado. Nesse caso, é impresso na saída as primeiras linhas do dataframe 
modificado, usando o método "head".
'''

df_trienio.set_index('NU_ANO_CENSO',
                    #  inplace=True
                     ).head()
```

```python
# Selecionando duas ou mais colunas do dataframe para ser o índice

'''
O código a seguir utiliza o método "set_index" no dataframe "df_trienio" com base 
nas colunas especificadas na lista "colunas". Isso significa que essas colunas 
serão removidas do conjunto de colunas do dataframe e colocadas como índice. No caso, 
as colunas selecionadas foram "NU_ANO_CENSO" e "NO_IES". A opção "inplace=True" não 
foi utilizada, o que significa que o resultado do set_index não será armazenado no 
próprio dataframe "df_trienio". O método "head" é utilizado para exibir as primeiras 
linhas do dataframe resultante.
'''

colunas = ['NU_ANO_CENSO', 'NO_IES']

df_trienio.set_index(colunas,
                    #  inplace=True
                     ).head()
```

```python
# Selecionando uma coluna como índice e adicionando também um índice numérico

'''

O código a seguir usa o método "set_index" para definir um novo índice para o 
dataframe "df_trienio". O novo índice é uma combinação de um intervalo de números 
inteiros de "1" até "n", onde "n" é o número de linhas do dataframe, e a coluna 
"NO_IES". Isso significa que o índice do dataframe terá duas colunas: a primeira 
é um número inteiro único para cada linha do dataframe e a segunda é o nome da 
instituição de ensino. Ao passar uma lista de dois elementos para a função "set_index¨, 
que contém a coluna NO_IES e uma lista de inteiros, a função "set_index" cria um 
MultiIndex no dataframe, tornando a pesquisa e a análise mais eficientes. O argumento 
"inplace=True" é opcional e, se definido como "True", a operação de reindexação é 
realizada diretamente no dataframe "df_trienio", em vez de criar um novo dataframe.
'''

n = df_trienio.shape[0]

df_trienio.set_index([pd.Index(range(1, n+1)), 
                      'NO_IES'], 
                    #  inplace=True
                     ).head()
```

## 6. Renomeando colunas

O comando rename do Python é uma função utilizada para renomear colunas de um dataframe. Com esse comando, é possível alterar o nome de uma ou mais colunas de um dataframe de forma rápida e simples. Além disso, é possível utilizar o comando para renomear índices do dataframe. Essa função é especialmente útil quando se trabalha com dados de diferentes fontes e é necessário padronizar os nomes das colunas para facilitar a análise e o processamento dos dados. Com o rename, é possível realizar essa padronização de forma eficiente e automatizada.

```python
# Renomear manualmente, uma por uma

'''
O código a seguir tem como objetivo renomear as colunas de um dataframe, sendo 
que o novo nome de cada coluna é especificado no dicionário "colunas". A função 
utilizada para realizar esta ação é a "rename", que recebe como argumentos o 
dicionário com os novos nomes de cada coluna e o parâmetro "inplace', que se 
verdadeiro, realiza a operação no próprio dataframe, sem criar uma cópia. Neste 
caso, a opção inplace não foi utilizada, então a função retorna uma cópia do 
dataframe original com as colunas renomeadas e armazenado na variável "df". As 
colunas renomeadas foram 'NU_ANO_CENSO', 'CO_IES' e 'NO_IES', que passaram a 
se chamar 'Ano', 'Código' e 'Instituição', respectivamente.
'''

colunas = {'NU_ANO_CENSO':'Ano',
           'CO_IES':'Código',
           'NO_IES':'Instituição'}

df = df_trienio.rename(columns=colunas,
                      #  inplace = True
                       )

df.head()
```

```python
# Renomeando todas as colunas em letra minúscula

'''
O código abaixo converte as colunas do dataframe df em letras minúsculas. A função 
"str.lower" é aplicada às colunas do dataframe, que contém os nomes das colunas em 
formato de string. Isso resulta em uma nova lista de colunas com todas as letras 
em minúsculo. Esse procedimento é útil em casos em que as colunas foram criadas 
com letras maiúsculas ou misturadas e se deseja padronizar os nomes das colunas 
para facilitar a manipulação e análise dos dados.
'''

df.columns = df.columns.str.lower()

df.head()
```

```python
# Somente a primeira letra de cada palavra em maiúscula

'''
Este código é usado para renomear todas as colunas do dataframe "df", transformando 
a primeira letra de cada palavra em maiúscula. A função "title" é aplicada a cada 
string do nome das colunas para executar essa transformação. Por exemplo, se uma 
coluna se chamava "nome_do_aluno", após a execução desse código, seu nome seria 
alterado para "Nome_Do_Aluno".
'''

df.columns = df.columns.str.title()

df.head()
```

```python
# Renomeando todas as colunas em letra maiúscula

'''
O código abaixo converte as colunas do dataframe df em letras maiúsculas. A função 
"str.lower" é aplicada às colunas do dataframe, que contém os nomes das colunas em 
formato de string. Isso resulta em uma nova lista de colunas com todas as letras 
em maiúsculo. Esse procedimento é útil em casos em que as colunas foram criadas 
com letras minúsculas ou misturadas e se deseja padronizar os nomes das colunas 
para facilitar a manipulação e análise dos dados.
'''

df.columns = df.columns.str.upper()

df.head()
```

```python
# Remover string specíficas dos nomes das colunas do dataframe

'''
O código abaixo consiste em um loop "for" que percorre uma lista de substrings que 
se deseja remover do nome das colunas de um dataframe. Para cada substring na lista, 
é aplicada uma função de substituição que remove a substring do nome das colunas, 
elimina espaços em branco extras e converte o texto resultante para letras maiúsculas. 
O loop percorre cada coluna do dataframe e a substituição é aplicada somente se a 
coluna é uma string e se contém a substring a ser removida. A nova lista de colunas 
resultante é armazenada de volta no dataframe. O objetivo dessa operação é limpar o 
nome das colunas, removendo prefixos e sufixos que podem atrapalhar a análise dos dados.
'''

list_strings = ['TP_', 
                'NO_', 
                'QT_', 
                'IN_', 
                'CO_']

for str_to_replace in list_strings:
    df.columns = [col.replace(str_to_replace, '').strip().upper() if isinstance(col, str) and str_to_replace in col else col for col in df.columns]

df.head()
```

```python
# Exclusão de caracteres especiais entre dos nomes de colunas

'''
O código abaixo tem como objetivo substituir todos os caracteres especiais nas colunas 
do dataframe "df" por espaços em branco. Para realizar esta tarefa, o método "str.replace"
é aplicado nas colunas do dataframe "df" usando a sintaxe "df.columns.str.replace()". O 
primeiro argumento da função "replace" é o caractere que se deseja substituir, e o 
segundo argumento é o caractere que será usado para substituir. Neste caso, o 
segundo argumento é um espaço em branco (' '). Ao final do código, o atributo "columns" 
do dataframe "df" é atualizado com as novas colunas que agora contêm espaços em branco em 
vez de '_'.
'''

df.columns=df.columns.str.replace('_',' ')

df.head()
```

## 8. Linhas duplicadas

A identificação de linhas duplicadas é uma etapa importante na análise de dados, e em Python, existem diversas maneiras de realizar essa tarefa. A detecção de linhas duplicadas pode ajudar a identificar erros nos dados, bem como evitar distorções em análises e resultados. Com o uso de funções e métodos específicos, é possível identificar e remover essas linhas duplicadas de forma eficiente e precisa, garantindo a qualidade e confiabilidade dos dados analisados.

```python
# Contagem de linhas duplicadas por coluna do dataframe

'''
O código define um valor para a opção "max_rows" do módulo Pandas, que 
é usada para exibir o número máximo de linhas de uma tabela de dados. O valor 
atribuído é o número de colunas do dataframe "df_trienio" acrescido de 1, o que 
significa que todas as linhas serão exibidas na tabela.

Em seguida, o código calcula o número de duplicatas para cada coluna do dataframe 
"df_trienio" usando o método "duplicated()" e a função "sum()" do Pandas. Essa 
contagem é armazenada na variável "column_duplicated_counts".

Por fim, o código cria um novo dataframe chamado "result_table" usando a função 
"pd.DataFrame()". Este dataframe contém duas colunas: "Column Name" (nome da coluna) 
e "Duplicate Count" (contagem de duplicatas). Os valores dessas colunas são extraídos 
da variável "column_duplicated_counts". Por fim, a função "display()" é usada para 
exibir a tabela "result_table" na saída. Em resumo, o código mostra o número de 
duplicatas em cada coluna do dataframe "df_trienio".
'''

pd.options.display.max_rows = df_trienio.shape[1] + 1

column_duplicated_counts = df_trienio[df_trienio.duplicated()].sum(axis=0)

result_table = pd.DataFrame({'Nome da coluna': column_duplicated_counts.index, 'Quantidade de duplicatas': column_duplicated_counts.values})

display(result_table)
```

```python
# Visualização das linhas duplicadas

'''
O código seleciona as linhas duplicadas no dataframe "df_trienio" 
utilizando o método "duplicated()" do Pandas. A variável "duplicado" armazena as 
linhas duplicadas do dataframe.

Em seguida, o código calcula o número de linhas duplicadas usando a função "len()" 
do Python e armazena o resultado na variável "num_duplicatas".

O código imprime o número de linhas duplicadas utilizando a função "print()" do 
Python, com uma mensagem informativa formatada com a variável "num_duplicatas". 
Em seguida, utiliza outra função "print()" para exibir a mensagem "Linhas Duplicadas" 
seguida de duas quebras de linha para criar uma linha vazia.

Por fim, a variável "duplicado" é exibida na saída. Essa variável contém as 
linhas duplicadas do dataframe "df_trienio", o que permite identificar as linhas 
específicas que estão duplicadas. Em resumo, o código mostra o número de linhas 
duplicadas no dataframe "df_trienio" e exibe as próprias linhas duplicadas.
'''

duplicado = df_trienio[df_trienio.duplicated()]

num_duplicatas = len(duplicado)

print(f"Número de linhas duplicadas: {num_duplicatas} \n")

print("Linhas Duplicadas: \n\n")

duplicado
```

```python
# Eliminando todas as duplicatas de todas as colunas

'''
O código utiliza o método "drop_duplicates()" do Pandas para 
remover as linhas duplicadas do dataframe "df_trienio". Este método retorna um 
novo dataframe com as linhas duplicadas removidas.

Em resumo, o código remove as linhas duplicadas do dataframe "df_trienio" e 
exibe as primeiras linhas do dataframe resultante sem as linhas duplicadas.

Obs.: Como não há duplicatas, nenhuma linha será removida.
'''

df_trienio.drop_duplicates().head()
```

```python
# Eliminando duplicatas de um subconjunto de colunas

'''
O código utiliza o método "drop_duplicates()" do Pandas para 
remover as linhas duplicadas do dataframe "df_trienio". O parâmetro "subset" é 
definido como ['NU_ANO_CENSO', 'CO_IES'], o que significa que as linhas 
duplicadas serão identificadas apenas considerando essas duas colunas. Esse 
método retorna um novo dataframe com as linhas duplicadas removidas.

Em resumo, o código remove as linhas duplicadas do dataframe "df_trienio", 
considerando apenas as colunas 'NU_ANO_CENSO' e 'CO_IES', e exibe as primeiras 
linhas do dataframe resultante sem as linhas duplicadas. Essa abordagem permite 
remover as linhas duplicadas com base em critérios específicos, considerando 
apenas as colunas relevantes para a análise e preservando as demais informações 
do dataframe.

Obs.: Como não há duplicatas, nenhuma linha será removida.
'''

df_trienio.drop_duplicates(subset=['NU_ANO_CENSO', 'CO_IES']).head()
```

Felizmente, Python oferece diversas funções para lidar com dados ausentes em dataframes, permitindo a remoção de linhas e colunas que contenham valores ausentes ou preenchendo esses valores com dados de outras colunas ou estatísticas do conjunto de dados. A eliminação de dados ausentes é uma prática essencial para garantir a integridade e a confiabilidade dos resultados da análise de dados.

```python
# Identificando o total de dados ausentes

'''
O código abaixo utiliza o método "isna()" do Pandas para identificar valores ausentes 
no dataframe "df_trienio". O método "isna()" retorna um dataframe booleano que 
indica se cada célula do dataframe é nula (True) ou não (False).

Em seguida, o método "sum()" é aplicado duas vezes no dataframe resultante. A 
primeira aplicação do método soma os valores nulos em cada coluna do dataframe, 
retornando um novo dataframe com o total de valores nulos em cada coluna. A segunda 
aplicação do método soma os valores nulos em todo o dataframe, retornando o 
número total de valores nulos em todo o dataframe.

Por fim, a função "print()" é utilizada para exibir a mensagem "Total de dados 
ausentes: " e o número total de valores nulos em todo o dataframe.

Em resumo, o código a seguir calcula e exibe o número total de valores nulos em todo 
o dataframe "df_trienio".
'''

print('Total de dados ausentes: ', df_trienio.isna().sum().sum())
```

## 10. Filtros

O filtro é uma técnica muito utilizada na análise de dados com Python, permitindo selecionar apenas os dados que atendem a determinadas condições. Com o uso de operadores lógicos e de comparação, é possível selecionar linhas e colunas do dataframe que atendam a critérios específicos, facilitando a análise de dados e a tomada de decisões baseadas em informações precisas. É uma ferramenta poderosa para lidar com grandes conjuntos de dados e facilitar a análise exploratória.

```python
# Filtrar dados por ano

'''
O código abaixo realiza a criação de um filtro para selecionar apenas as linhas do 
dataframe df_trienio que possuem o valor "2009" na coluna "NU_ANO_CENSO". Esse filtro
é criado por meio da comparação df_trienio['NU_ANO_CENSO'] == 2009. Em seguida, o 
filtro é aplicado no dataframe original utilizando o método loc[], criando assim 
um novo dataframe chamado df_2009 com as linhas filtradas. Portanto, o resultado 
é um dataframe que contém apenas as informações referentes ao ano de 2009.
'''

filtro_ano = (df_trienio['NU_ANO_CENSO'] == 2009)

df_2009 = df_trienio.loc[filtro_ano]

df_2009.head(2)
```

```python
# Filtrar dados por categoria administrativa da IES

'''
O código a seguir cria um filtro para selecionar apenas as linhas do dataframe 
"df_trienio" que possuem o valor 'Pública Federal' na coluna 'NO_CATEGORIA_ADMINISTRATIVA'. 
Isso é feito por meio da criação de uma variável chamada "filtro_categ_adm" que contém 
a condição lógica para selecionar essas linhas.

Em seguida, o filtro é aplicado ao dataframe "df_trienio" utilizando o método "loc[]" 
do Pandas. Esse método permite selecionar apenas as linhas do dataframe que atendem à 
condição especificada no filtro. O resultado da filtragem é atribuído a uma nova 
variável chamada "df_categ_adm".
'''

filtro_categ_adm = (df_trienio['NO_CATEGORIA_ADMINISTRATIVA'] == 'Pública Federal')

df_categ_adm = df_trienio.loc[filtro_categ_adm]

df_categ_adm.head(2)
```

```python
# Filtrar por ano e categoria administrativa da IES

'''
O código abaixo cria um filtro para selecionar apenas as linhas do dataframe "df_trienio" 
que possuem o valor 'Pública Federal' na coluna 'NO_CATEGORIA_ADMINISTRATIVA' e o 
valor '2011' na coluna 'NU_ANO_CENSO'. Isso é feito por meio da criação de uma variável 
chamada "filt_categ_adm_2011" que contém a condição lógica para selecionar essas linhas.

O operador "\" no final da primeira linha é usado para dividir a condição lógica em duas 
linhas para melhor legibilidade do código.

Em seguida, o filtro é aplicado ao dataframe "df_trienio" utilizando o operador [] do 
Pandas. Esse operador permite selecionar apenas as linhas do dataframe que atendem à 
condição especificada no filtro. O resultado da filtragem é atribuído a uma nova variável 
chamada "df_categ_adm_2011".
'''

filt_categ_adm_2011 = ((df_trienio['NO_CATEGORIA_ADMINISTRATIVA'] == 'Pública Federal') \
                       & (df_trienio['NU_ANO_CENSO'] == 2011))

df_categ_adm_2011 = df_trienio[filt_categ_adm_2011]

df_categ_adm_2011.head(2)
```

```python
# Filtrar por ano, categoria administrativa da IES e nome da organização acadêmica

'''
O código abaixo cria um filtro para selecionar apenas as linhas do dataframe "df_trienio" 
que possuem o valor 'Pública Federal' na coluna 'NO_CATEGORIA_ADMINISTRATIVA', o valor 
'2011' na coluna 'NU_ANO_CENSO' e o valor 'Universidade' na coluna 'NO_ORGANIZACAO_ACADEMICA'. 
Essa operação é realizada por meio da criação de uma variável chamada "filt_categ_adm_nome_2011", 
que contém a condição lógica para selecionar essas linhas.

O operador "\" é usado para dividir a condição lógica em três linhas para melhor legibilidade do código.

Em seguida, o filtro é aplicado ao dataframe "df_trienio" utilizando o operador [] do Pandas. 
Esse operador permite selecionar apenas as linhas do dataframe que atendem à condição especificada 
no filtro. O resultado da filtragem é atribuído a uma nova variável chamada "df_categ_adm_nome_2011".
'''

filt_categ_adm_nome_2011 = ((df_trienio['NO_CATEGORIA_ADMINISTRATIVA'] == 'Pública Federal') \
                            & (df_trienio['NU_ANO_CENSO'] == 2011) \
                            & (df_trienio['NO_ORGANIZACAO_ACADEMICA'] == 'Universidade'))

df_categ_adm_nome_2011 = df_trienio[filt_categ_adm_nome_2011]

df_categ_adm_nome_2011.head(2)
```

```python
# Filtrar por nome da organização acadêmica e nome da categoria administrativa

'''
Este código realiza uma filtragem do dataframe df_trienio para retornar as linhas 
que possuem os valores 'Universidade' na coluna 'NO_ORGANIZACAO_ACADEMICA' e 
'Privada sem fins lucrativos' ou 'Privada com fins lucrativos' na coluna 
'NO_CATEGORIA_ADMINISTRATIVA'.

O operador "\" é usado para dividir a condição lógica em três linhas para melhor 
legibilidade do código.

Primeiro, o filtro é definido na variável filtro_nome_org_categ_adm através de uma 
expressão booleana. Em seguida, é criado o dataframe df_nome_org_categ_adm através 
da aplicação desse filtro ao dataframe original. A variável df_nome_org_categ_adm 
contém somente as linhas que atendem às condições especificadas pelo filtro.
'''

filtro_nome_org_categ_adm = ((df_trienio['NO_ORGANIZACAO_ACADEMICA'] == 'Universidade') \
                              & ((df_trienio['NO_CATEGORIA_ADMINISTRATIVA'] == 'Privada sem fins lucrativos') \
                              | (df_trienio['NO_CATEGORIA_ADMINISTRATIVA'] == 'Privada com fins lucrativos')))


df_nome_org_categ_adm = df_trienio[filtro_nome_org_categ_adm]

df_nome_org_categ_adm.head(2)
```

**11.2 Outliers**

Outliers são valores que se afastam significativamente do restante dos dados em um conjunto de dados. Eles podem ser resultado de erros de medição ou representar valores extremos dentro da população em análise. Identificar outliers é importante para entender a distribuição dos dados e tomar decisões informadas em análises estatísticas. Existem diversas técnicas para identificar outliers, como a análise de boxplot, o cálculo de desvio padrão e o método interquartil. 

```python
# Identificação das linhas com outliers

'''
Este código identifica outliers na coluna "QT_VAGAS_ANUAL_NOTURNO" do dataframe 
"df_trienio". Primeiramente, a variável "coluna" recebe o nome da coluna que será 
analisada. Em seguida, é calculado o primeiro quartil (Q1), o terceiro quartil 
(Q3) e o intervalo interquartil (IQR) para a coluna especificada, utilizando a 
função "quantile" do Pandas. Com base nesses valores, são calculados o limite 
superior e o limite inferior para os dados da coluna, sendo o limite superior 
calculado através da soma de Q3 com 1,5 vezes o IQR, e o limite inferior calculado 
pela subtração de Q1 com 1,5 vezes o IQR. A função "where" do Numpy é utilizada 
para identificar as linhas do dataframe que contêm valores maiores que o limite 
superior, e os índices dessas linhas são armazenados na variável "outlier_linhas". 
Por fim, é exibida na tela uma mensagem contendo a lista de índices das linhas do 
dataframe que contêm outliers para a coluna especificada. Esse tipo de código é 
útil para identificar e tratar outliers em colunas específicas, garantindo uma 
análise mais precisa dos dados.
'''

coluna = 'QT_VAGAS_ANUAL_NOTURNO'

Q1 = df_trienio[coluna].quantile(0.25)

Q3 = df_trienio[coluna].quantile(0.75)

IQR = Q3 - Q1

limite_superior = Q3 + (1.5 * IQR)
                      
limite_inferior = Q1 - (1.5 * IQR)                        

outlier_linhas = np.where(df_trienio[coluna] > limite_superior)[0]

print('Linhas com outliers para a', coluna, 'do dataframe df_trienio: \n\n', outlier_linhas)
```

```python
# Número de outliers por coluna que inicia com "QT_"

'''
Este código conta o número de outliers em cada coluna que começa com "QT_" do 
dataframe "df_trienio".

Primeiro, as colunas que começam com "QT_" são selecionadas usando o método "filter" 
e armazenadas na variável "df". Em seguida, é definida uma função chamada "contar_outliers" 
que recebe uma coluna como entrada. Dentro dessa função, a coluna é filtrada para 
remover os valores igual a zero (zeros podem ser usados para preencher campos vazios). 
Em seguida, é calculado o primeiro e terceiro quartil (Q1 e Q3) da coluna e, a partir 
disso, é calculado o intervalo interquartil (IQR) e o limiar para identificar outliers 
(limiar é definido como 1,5 vezes o IQR). Por fim, é retornado o número de valores 
que são identificados como outliers.

Na sequência, a função "contar_outliers" é aplicada a todas as colunas de "df" usando 
o método "apply" e o resultado é armazenado na variável "outliers_por_coluna". Por fim, 
é feita uma impressão com o número de outliers encontrados por coluna.
'''

df = df_trienio.filter(regex='^QT_')

def contar_outliers(coluna):
    coluna = coluna[coluna != 0] 
    if len(coluna) == 0: 
        return 0
    q1, q3 = np.percentile(coluna, [25, 75])
    iqr = q3 - q1
    limiar = 1.5 
    
    outliers = (coluna < q1 - limiar*iqr) | (coluna > q3 + limiar*iqr)
    return np.sum(outliers)

outliers_por_coluna = df.apply(contar_outliers)

outliers_por_coluna = outliers_por_coluna[outliers_por_coluna != 0]

print('Número de outliers por coluna:\n\n', outliers_por_coluna)
```

```python
# Intervalo Interquartil

'''
Este código calcula o intervalo interquartil (IQR) para todas as colunas do DataFrame 
"df_trienio" que iniciam com os caracteres "QT_", utilizando as funções "filter" e 
"quantile" do Pandas. Primeiramente, a variável "Q1" é criada para armazenar o primeiro 
quartil (25%) dos dados de todas as colunas que atendem ao filtro. Em seguida, a 
variável "Q3" é criada para armazenar o terceiro quartil (75%) dos mesmos dados. 
Com base nesses valores, é possível calcular o IQR para todas as colunas utilizando 
a subtração de Q1 de Q3. A variável "non_zero_IQR" é criada para armazenar apenas os 
valores de IQR que são diferentes de zero, filtrando as colunas que não possuem 
variação nos dados. Por fim, é exibido na tela o valor de "non_zero_IQR", que contém 
uma lista dos valores de IQR para as colunas que possuem variação nos dados. Esse 
tipo de código é útil para identificar as colunas que possuem uma variação significativa 
nos dados, permitindo uma análise mais precisa e focada nessas colunas específicas.
'''

Q1 = df_trienio.filter(regex='^QT_').quantile(0.25)

Q3 = df_trienio.filter(regex='^QT_').quantile(0.75)

IQR = Q3 - Q1

nao_zero_IQR = IQR[IQR != 0]

print(nao_zero_IQR)
```

```python
# Remoção de outliers - Apenas graficamente

'''
Este código plota um boxplot da variável "QT_VAGAS_ANUAL_NOTURNO" do DataFrame 
"df_trienio" usando a biblioteca Seaborn. O parâmetro "x" especifica que a variável 
"QT_VAGAS_ANUAL_NOTURNO" deve ser usada no eixo horizontal do gráfico. O parâmetro 
"showfliers" é definido como "False", o que significa que os outliers serão excluídos 
do boxplot. O comando "plt.xlabel" adiciona um rótulo ao eixo horizontal do gráfico 
e "plt.show" exibe o gráfico.
'''

sns.boxplot(x=df_trienio['QT_VAGAS_ANUAL_NOTURNO'], showfliers=False)

plt.xlabel('\n Vagas Anual Noturno')

plt.show()
```

```python
# Remoção de outliers - No dataframe

'''
Este código filtra o dataframe df_trienio para incluir apenas as colunas 
que começam com 'QT_'. Em seguida, calcula o primeiro quartil (Q1), o terceiro 
quartil (Q3) e a amplitude interquartil (IQR) para cada uma dessas colunas. O 
limite inferior (lim_inferior) e superior (lim_superior) para identificar os outliers 
são calculados com base no IQR. Depois, um novo dataframe é criado (df_sem_outliers)
com todas as linhas que não contêm nenhum outlier para as colunas selecionadas. 
Este novo dataframe é retornado como resultado. A função "all(axis=1)" é usada para 
garantir que as linhas com outliers em qualquer coluna sejam removidas.
'''

df = df_trienio.filter(regex='^QT_')

Q1 = df.quantile(0.25)

Q3 = df.quantile(0.75)

IQR = Q3 - Q1
 
lim_inferior = Q1 - 1.5 * IQR

lim_superior = Q3 + 1.5 * IQR

df_sem_outliers = df.loc[((df >= lim_inferior) & (df <= lim_superior)).all(axis=1)]

df_sem_outliers.head()
```

```python
# Diagrama de caixas após a remoção dos outliers no dataframe

'''
Este código gera um boxplot da coluna "QT_VAGAS_ANUAL_NOTURNO" do dataframe 
"df_sem_outliers", utilizando a biblioteca Seaborn.
'''

sns.boxplot(x=df_sem_outliers['QT_VAGAS_ANUAL_NOTURNO'])

plt.xlabel('\n Vagas Anual Noturno')

plt.show()
```

```python
# Quantidade de linhas removidas

'''
O código abaixo cria dois dataframes: df_1 e df_2, e depois imprimir informações 
sobre eles, como o número de linhas e colunas em cada um. O dataframe df_1 é gerado 
a partir do dataset original (df_trienio) e inclui todas as colunas que começam com 
'QT_'. O dataframe df_2 é gerado a partir do dataframe df_1 após a remoção dos outliers 
das colunas incluídas em df_1. A impressão na tela apresenta informações sobre ambos 
os dataframes, incluindo o número de linhas e colunas em cada um, e a diferença no 
número de linhas entre df_1 e df_2, que é o número de linhas que foram removidas 
na etapa de remoção de outliers.
'''

df_1 = df_trienio.filter(regex='^QT_')

df_2 = df_sem_outliers

print('df_1 (com outliers): \n\n Linhas =', 
      df_1.shape[0], 
      '\n Colunas =', 
      df_1.shape[1],
      '\n\n df_2 (sem outliers): \n\n Linhas =', 
      df_2.shape[0], 
      '\n Colunas =', 
      df_2.shape[1],
      '\n\n Linhas removidas =',
      (df_1.shape[0] - df_2.shape[0]))
```

Eles são uma ferramenta eficaz para visualizar rapidamente as informações e identificar tendências ou padrões, permitindo que tomadas de decisão sejam mais assertivas.

```python
# Gráfico de barras de uma variável

'''
Este código realiza o cálculo do total de concluintes por ano, agrupando os dados 
do dataframe df_trienio pela coluna 'NU_ANO_CENSO' e somando a quantidade de 
concluintes. Em seguida, é determinado o valor máximo para o eixo y, considerando 
os totais de concluintes, multiplicado por 1.1 para adicionar uma margem. A coluna 
de índices é convertida de tipo para que seja exibida como strings no eixo x. O 
gráfico de barras é plotado, utilizando os índices no eixo x e os valores dos 
totais de concluintes no eixo y. O eixo x é rotulado como "Ano" e o eixo y como 
"Total de concluintes". O limite superior do eixo y é ajustado de acordo com o 
valor máximo calculado. O formato do rótulo do eixo y é definido como "plain" 
para exibir números inteiros. Por fim, o gráfico é exibido utilizando plt.show().
'''

soma_concluintes_por_ano = df_trienio.groupby('NU_ANO_CENSO')['QT_CONCLUINTE_TOTAL'].sum()

max_y = max(soma_concluintes_por_ano.values) * 1.1

soma_concluintes_por_ano.index = soma_concluintes_por_ano.index.astype(int).astype(str)

fig, ax = plt.subplots()
ax.bar(soma_concluintes_por_ano.index, soma_concluintes_por_ano.values)
ax.set_xlabel('Ano')
ax.set_ylabel('Total de concluintes')
# ax.set_title('Total de concluintes por ano')

ax.set_ylim(0, max_y)
ax.ticklabel_format(axis='y', style='plain')
plt.show()
```

### Barras empilhadas

Gráficos de barras empilhadas são uma variação dos gráficos de barras tradicionais, que permitem visualizar a composição dos dados de uma maneira mais detalhada. Em um gráfico de barras empilhadas, cada barra é dividida em segmentos coloridos que representam diferentes categorias ou subcategorias de dados. Esses segmentos são empilhados uns sobre os outros, permitindo que se visualize a contribuição de cada categoria para o total representado pela barra. Essa técnica é útil para comparar a distribuição de categorias em diferentes grupos de dados.

```python
# Gráfico de barras empilhado

'''
O código a seguir realiza a plotagem de um gráfico de barras para exibir os totais 
de concluintes e ingressantes por ano. Primeiramente, é feito um agrupamento dos 
dados pelo ano censitário e calculada a soma dos valores de concluintes e ingressantes 
para cada ano. Em seguida, é determinado o valor máximo para o eixo y, considerando 
os totais de ingressantes e concluintes, multiplicado por 1.1 para adicionar uma 
margem. O índice do DataFrame é convertido para uma representação de string. A 
partir disso, é criada uma figura e um conjunto de eixos. A variável x é criada 
para representar a posição das barras no eixo x. As barras de ingressantes e 
concluintes são desenhadas usando os dados correspondentes, e são atribuídas as 
cores laranja e azul, respectivamente. O eixo x é rotulado como "Ano" e as legendas 
são adicionadas. O limite do eixo y é definido com base no valor máximo calculado 
anteriormente. O formato dos rótulos do eixo y é definido como um número inteiro. 
Os rótulos do eixo x são definidos com base nos valores do índice do DataFrame. 
Finalmente, o gráfico é exibido.
'''

soma_por_ano = df_trienio.groupby('NU_ANO_CENSO')[['QT_CONCLUINTE_TOTAL', 'QT_INGRESSO_TOTAL']].sum()
max_y = max(soma_por_ano[['QT_INGRESSO_TOTAL', 'QT_CONCLUINTE_TOTAL']].values.flatten()) * 1.1

soma_por_ano.index = soma_por_ano.index.astype(str)  

fig, ax = plt.subplots()
x = np.arange(len(soma_por_ano.index))  
ax.bar(x, soma_por_ano['QT_INGRESSO_TOTAL'], label='Ingressantes', color='orange')
ax.bar(x, soma_por_ano['QT_CONCLUINTE_TOTAL'], label='Concluintes', color='blue')
ax.set_xlabel('Ano')
# ax.set_ylabel('Total')
# ax.set_title('Total de concluintes e ingressantes por ano')

ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2)

ax.set_ylim(0, max_y)
ax.ticklabel_format(axis='y', style='plain')
ax.set_xticks(x)  
ax.set_xticklabels(soma_por_ano.index) 

plt.show()
```

### Mapas de calor

Gráficos do tipo mapa de calor são uma representação visual de dados que utiliza cores para indicar a magnitude de uma variável em relação a outras. Geralmente, são utilizados em análises de dados multidimensionais, em que é necessário identificar padrões ou correlações entre variáveis. O mapa de calor é construído a partir de uma matriz de dados, em que as linhas representam uma variável e as colunas representam outra. Cada célula da matriz é preenchida com uma cor que representa o valor da variável em relação às outras. Quanto mais escuro o tom da cor, maior o valor da variável, e quanto mais claro, menor o valor.

```python
# Mapa de calor

'''
Este código cria um heatmap da correlação entre algumas variáveis do dataframe 
df_trienio, incluindo QT_MATRICULA_TOTAL, QT_CONCLUINTE_TOTAL, QT_INGRESSO_PROCESSO_SELETIVO, 
QT_VAGA_TOTAL e QT_INSCRITO_TOTAL. Para isso, é utilizada a função corr() do pandas 
para calcular a correlação entre essas variáveis e em seguida, a biblioteca Seaborn 
é utilizada para plotar o heatmap. A intensidade da cor indica a força e direção 
da correlação, enquanto que a escala de cores azul e vermelho indica a direção da 
correlação (negativa ou positiva, respectivamente). A opção annot=True adiciona os 
valores de correlação dentro de cada célula do heatmap.
'''

plt.figure(figsize=(10,6))

corr = df_trienio[['QT_MATRICULA_TOTAL', 
                   'QT_CONCLUINTE_TOTAL', 
                   'QT_INGRESSO_PROCESSO_SELETIVO', 
                   'QT_VAGA_TOTAL', 
                   'QT_INSCRITO_TOTAL']].corr()

sns.heatmap(corr,
            cmap='coolwarm',
            annot=True
            );

plt.show()
```

Cada setor representa uma categoria e seu tamanho é proporcional à porcentagem que essa categoria representa em relação ao todo. ### Anel

Gráficos de anel, também conhecidos como gráficos de rosca, são uma variação dos gráficos de pizza que exibem dados em um formato circular. Eles mostram as partes de um todo em porcentagens, sendo que cada seção do anel representa uma categoria específica. Em contraste com os gráficos de pizza, os gráficos de anel possuem um espaço vazio no centro, o que pode ser usado para destacar uma categoria ou uma informação adicional. 

```python
# Triênio - 2009, 2010 e 2011

'''
Este código gera um gráfico de pizza para exibir a distribuição da categoria 
administrativa das instituições de ensino superior no triênio 2009-2010-2011. 
Primeiramente, a variável 'contagem' é criada para armazenar a contagem de 
instituições em cada categoria administrativa. Em seguida, é criada uma figura 
com um único subplot de tamanho 8x8 e o gráfico de pizza é gerado usando o método 
'pie' do objeto 'ax'. A porcentagem de cada categoria é exibida no gráfico por meio 
do parâmetro 'autopct'. O parâmetro 'radius' define o tamanho do gráfico de pizza. 
O parâmetro 'wedgeprops' permite customizar a espessura das fatias do gráfico e o 
parâmetro 'edgecolor' define a cor da borda das fatias. Em seguida, é criado um 
círculo no centro do gráfico com um raio menor que o gráfico de pizza, o que 
resulta em um "buraco" no centro do gráfico. O círculo é adicionado à figura 
com o método 'add_artist'. Por fim, é definido um título para o gráfico usando 
o método 'set_title'.
'''

contagem = df_trienio['NO_CATEGORIA_ADMINISTRATIVA'].value_counts()

fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(contagem.values, labels=contagem.index, autopct='%1.1f%%', radius=1, wedgeprops=dict(width=0.3, edgecolor='w'))

centro_circulo = plt.Circle((0,0),0.7,color='black', fc='white',linewidth=1.25)
fig = plt.gcf()
fig.gca().add_artist(centro_circulo)

ax.set_title('Categoria Administrativa das IES no Triênio 2009-2010-2011')
plt.show()
```

### Histograma

Um histograma é um gráfico de frequência que representa a distribuição de valores de uma variável quantitativa. Ele é construído a partir da divisão do intervalo dos valores da variável em várias faixas, chamadas de "bins", e contabiliza quantos valores da variável estão dentro de cada bin. Assim, o eixo x do histograma representa os valores da variável, enquanto o eixo y representa a contagem de observações dentro de cada bin. O histograma é útil para visualizar a forma da distribuição dos dados, bem como a presença de eventuais assimetrias, outliers ou outras características relevantes.

```python
# Histograma do total de ingressantes por IES

'''
Este código cria um histograma que representa a distribuição da quantidade total 
de ingressantes por instituição de ensino superior (IES) em um período de três 
anos. Primeiramente, é feito um agrupamento dos dados do dataframe "df_trienio" 
pela coluna "CO_IES" (código da IES) e a soma da quantidade de ingressantes de cada 
uma delas é obtida. Em seguida, é definido o número de "bins" (intervalos) para 
o histograma e a figura é criada com o tamanho de 10x6. O histograma é gerado 
utilizando a função "hist" do Matplotlib, que recebe como parâmetro os valores 
do agrupamento, o número de intervalos e a cor laranja. São definidos os rótulos 
dos eixos x e y, limites dos eixos x e y, bem como a escala da marcação do eixo x. 
São calculados os percentuais de ocorrência de cada intervalo e é verificado se 
algum intervalo possui mais de 2% de ocorrência. Caso haja, é adicionado ao gráfico 
um texto com a porcentagem. Também é desenhada uma linha vermelha que representa 
a mediana da quantidade de ingressantes por IES.
'''

grupo = df_trienio.groupby('CO_IES')['QT_INGRESSO_TOTAL'].sum()

bins = 100

fig, ax = plt.subplots(figsize=(10, 6))

hist, edges, _ = ax.hist(grupo.values, bins=bins, color='orange')

ax.set_xlabel('\nIngressantes por ano no triênio 2009-2011')
ax.set_ylabel('Quantidade de IESs\n')

ax.set_xlim([0, df_trienio['QT_INGRESSO_TOTAL'].max()])
ax.set_xticks([i*1000 for i in range(int(df_trienio['QT_INGRESSO_TOTAL'].max()/1000)+1)]) 
ax.ticklabel_format(style='sci', axis='x', scilimits=(3,3), useMathText=True)

ylim = max(hist)
ax.set_ylim([0, ylim + 0.15 * ylim])

total = len(grupo)
percentages = (hist / total) * 100
max_index = np.argmax(hist)

if percentages[max_index] >= 2:
    ax.text(edges[max_index]+(edges[max_index+1]-edges[max_index])/2,\
            hist[max_index], f'{percentages[max_index]:.1f}%', ha='left',\
            va='bottom', color='green')

median = np.median(grupo.values)
ax.axvline(median, color='red')
ax.text(median*1.05, ylim*1.1, ' Mediana', color='red')

plt.show()
```

```python
# Dataframes - todos os arquivos de um diretório

'''
O código a seguir usa a biblioteca glob para buscar todos os arquivos CSV no diretório 
atual especificado pelo caminho "./". O comando "glob.glob" é usado para buscar todos
os arquivos que correspondem ao padrão especificado no caminho, que neste caso é "*.csv", 
ou seja, todos os arquivos com extensão CSV. A biblioteca "os.path" é usada para unir 
o caminho e o padrão de busca em um único caminho. Em seguida, o código usa um loop "for" 
para baixar cada um dos arquivos encontrados. O comando "gc.files.download" é usado 
para baixar cada arquivo encontrado pelo loop "for". Dessa forma, o código permite 
que o usuário baixe todos os arquivos CSV em um diretório específico. É importante 
notar que o comando "gc.files.download" é específico para o ambiente do Google Colab e 
não pode ser executado em outros ambientes de programação.
'''

caminho = './'

csv_arquivo = glob.glob(os.path.join(caminho, '*.csv'))

for arquivo in csv_arquivo:
    gc.files.download(arquivo)
```

```python
# Dataframes - todos os arquivos de um diretório compactados

'''
O código abaixo tem como objetivo compactar os arquivos CSV presentes no diretório 
especificado em um arquivo ZIP. Primeiro, é utilizado o módulo glob para obter 
uma lista de todos os arquivos com extensão .csv no diretório. Em seguida, é
definido o nome do arquivo ZIP como "arquivos_csv.zip".
Dentro do bloco with, um arquivo ZIP é criado utilizando a biblioteca zipfile. 
Em um loop, cada arquivo CSV é adicionado ao arquivo ZIP utilizando o método write() 
do objeto zipf. O nome de cada arquivo dentro do arquivo ZIP é definido utilizando 
a função os.path.basename(), garantindo que apenas o nome do arquivo seja utilizado.
Por fim, é feito o download do arquivo ZIP compactado utilizando a função 
files.download() do Google Colab, permitindo que o usuário faça o download do 
arquivo resultante contendo os arquivos CSV compactados em um único arquivo ZIP. 
Isso facilita o compartilhamento e transporte dos arquivos CSV em uma única 
unidade compactada.
'''

caminho = './'
arquivos_csv = glob.glob(os.path.join(caminho, '*.csv'))
arquivo_zip = 'arquivos_compactados.zip'

with zipfile.ZipFile(arquivo_zip, 'w') as zipf:
   
    for arquivo in arquivos_csv:
        zipf.write(arquivo, os.path.basename(arquivo))


gc.files.download(arquivo_zip)
```

```python
# Dataframes - apenas alguns arquivos de um diretório compactados

'''
O código a seguir tem como objetivo compactar apenas os arquivos indicados na lista 
arquivos em um arquivo ZIP chamado "arquivos_compactados.zip".
Dentro do bloco with, é criado um arquivo ZIP utilizando a biblioteca zipfile. 
Em um loop, cada arquivo da lista arquivos é verificado se existe no diretório 
especificado usando a função os.path.isfile(). Caso o arquivo exista, ele é 
adicionado ao arquivo ZIP utilizando o método write() do objeto zipf, com o 
nome do arquivo mantido.
Por fim, é feito o download do arquivo ZIP compactado utilizando a função 
files.download() do Google Colab, permitindo que o usuário faça o download 
do arquivo resultante contendo apenas os arquivos indicados na lista 
pré-fornecida. Isso facilita o compartilhamento e transporte específico 
desses arquivos em um único arquivo compactado.
'''

caminho = './'

arquivos = [
            'SoU_censo_IES_2009.csv', 
            'SoU_censo_IES_2011.csv', 
            'SoU_censo_IES_2010.csv', 
            'trienio_2009_2010_2011.csv'
           ]

arquivo_zip = 'arquivos_compactados.zip'

with zipfile.ZipFile(arquivo_zip, 'w') as zipf:
    for arquivo in arquivos:
        caminho_arquivo = os.path.join(caminho, arquivo)
        if os.path.isfile(caminho_arquivo):
            zipf.write(caminho_arquivo, arquivo)

gc.files.download(arquivo_zip)
```
